```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_dataset(file_path):
    """
    Performs data analysis and visualization on the Iris dataset.
    """
    try:
        # Task 1: Load and Inspect the Data
        df = pd.read_csv('/content/iris.csv')

        print("First 5 rows of the dataset:")
        print(df.head())

        print("\nData types and missing values:")
        print(df.info())

        # Clean the dataset (Iris dataset is clean, but showing the step)
        df.dropna(inplace=True)

        print("\nSuccessfully generated and saved all plots.")

    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    analyze_dataset('iris.csv')
```

Task 3: Data Visualization
Line Chart

```python
# Line chart (simulated time-series)
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['sepal_length'], label='Sepal Length')
plt.plot(df.index, df['petal_length'], label='Petal Length')
plt.title('Sepal and Petal Length Over Index')
plt.xlabel('Index')
plt.ylabel('Length (cm)')
plt.legend()
plt.savefig('line_chart.png')
plt.show()
plt.close()
```

Bar Chart

```python
# Bar chart
df = pd.read_csv('/content/iris.csv')
plt.figure(figsize=(10, 5))
sns.barplot(x='species', y='petal_length', data=df)
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.savefig('bar_chart.png')
plt.show()
plt.close()
```

Histogram

```python
# Histogram
plt.figure(figsize=(10, 5))
plt.hist(df['sepal_width'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.savefig('histogram.png')
plt.show()
plt.close()
```
